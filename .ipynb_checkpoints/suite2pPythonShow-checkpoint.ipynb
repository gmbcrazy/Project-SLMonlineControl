```python
meanImgE=ops['meanImgE']
#plt.imshow(meanImgE, cmap='viridis') 
fig, ax = plt.subplots(1,2)
ax[0].imshow(meanImgE, cmap='viridis')
#plt.plot(stat[3]['xpix'],stat[3]['ypix'],'r.')

im = np.zeros((ops['Ly'], ops['Lx']))
ncells=len(stat)
for n in range(0,ncells):
    if iscell[n,1]>0.5:
       #ypix = stat[n]['ypix'][~stat[n]['overlap']]
       #xpix = stat[n]['xpix'][~stat[n]['overlap']]        
       ypix = stat[n]['ypix']
       xpix = stat[n]['xpix']
       im[ops['Ly']-ypix,xpix] = n+1
    
ax[1].imshow(im,cmap=NeuroColor)
#plt.imshow(im)
#plt.show()
```

```python
import matplotlib.colors as mcolors

# Define a list of distinct colors
distinct_colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan']

# Specify the number of categories
num_categories = 100

# Create a colormap with repeated distinct colors
repeated_colors = [distinct_colors[i % len(distinct_colors)] for i in range(num_categories)]
NeuroColor = mcolors.ListedColormap(repeated_colors)
```
