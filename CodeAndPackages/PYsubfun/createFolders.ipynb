```python
import os
import pandas as pd
import shutil
import re
from classesandfunctions import Animal, getTseriesList
from pathlib import Path
#Make sure to change folder path to correct drive that you want folders to be created
folderpath = "/Volumes/Processed/"
```

```python
#This cell creates folders to move the correct Tseries files into

#Read Excel File and Print Column Headers
#Can change number of rows and skip rows in read excel function
specific_rows = [*range(5)]
files = pd.read_excel(
    '2023_Dev_Chronic_Summary.xlsx', 
    skiprows = lambda x: x not in specific_rows,
)
files = files.dropna(how='all')
files.head()

listOfAnimals = []

for index, row in files.iterrows():
    if(pd.isnull(row['DOB']) or pd.isnull(row['Date'])):
        pass
    else:
        dateOfBirth = row['DOB'][:8]
        animal = row['Animal']
        tSeries = row['Tseries']
        magnification = row['Mag']
        age = row['Date_2']
        recordingDate = row['Date']
        recordingStart = row['FileTime']
        listOfAnimals.append(Animal(dateOfBirth,animal,tSeries,magnification,age,recordingDate,recordingStart))


tiffFolders = []

for animal in listOfAnimals:
    birth = str(animal.recordingDate)
    name = str(animal.name)
    age = str(animal.age)
    
    folder = birth[:8] + "_" + name + "_" + age.lower()
    #Prints each folder name in list, double check list is correct
    #May Comment out line below if desired
    #print(folder)
    tiffFolders.append(folder)


for i in tiffFolders:
    filepath = Path(folderpath + "/" + i)
    if os.path.isdir(filepath):
        print("Directory: " + str(filepath) + " already exists")
        pass
    else:
        os.mkdir(filepath)
```
